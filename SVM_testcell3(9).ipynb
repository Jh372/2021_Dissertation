```python
def split(split_ratio, img_src):
    # get file list
    files = glob(join(img_src, '*.csv'))
    # split data into train and test sets(8:2)
    train_file, test_file = train_test_split(files, train_size=train_ratio)
    for f in range(len(train_file)):
        train_file[f] = os.path.split(train_file[f])[1]
    for f in range(len(test_file)):
        test_file[f] = os.path.split(test_file[f])[1]
    return train_file, test_file

# def train_SVM(filename, src):
```

```python
%load_ext email_notify_magic
%email hjunichi11@gmail.com --s 'Finish ML'
%email hjunichi11@gmail.com --body 'Check result' -k

test_file = np.loadtxt('./test_file9_list.csv', delimiter=',',skiprows=1, dtype='object')
train_file = np.loadtxt('./train_file9_list.csv', delimiter=',',skiprows=1, dtype='object')

# apply SVM
for num in range(len(train_file)):
    df = pd.read_csv(train_file[num])
#     df = np.loadtxt(train_file[num], delimiter=',', skiprows=1)
    clf = svm.SVC()
#     train_X,train_y = df[:,:-1],df[:,-1]
    train_X,train_y = df.iloc[:,:-1].values,df.iloc[:,-1].values
    clf.fit(train_X,train_y)
# output model
joblib.dump(clf, './train9.learn')
# load model
clf4 = joblib.load('./train9.learn')
# load test data
savename = join('./','./evaluation9_9.xlsx')
if os.path.exists(savename):
    os.remove(savename)
for num in range(len(test_file)):
#     df = pd.read_csv(train_file[num])
    df = np.loadtxt(test_file[num], delimiter=',', skiprows=1)
    test_X,test_y = df[:,:-1],df[:,-1]
    # result
    predicted_y = clf4.predict(test_X)
    # evaluation
    report = classification_report(test_y, predicted_y, output_dict=True)
    sheetname = os.path.splitext(os.path.basename(test_file[num]))[0]
    result = pd.DataFrame(report).transpose()
    if os.path.exists(savename):
        with pd.ExcelWriter(savename, engine="openpyxl", mode="a") as writer:
            result.to_excel(writer, sheet_name=sheetname)
    else:
        with pd.ExcelWriter(savename, engine="openpyxl") as writer:
            result.to_excel(writer, sheet_name=sheetname)
    

#     score = metrics.accuracy_score(test_y, predicted_y)
#     print("Score:", score)
```
